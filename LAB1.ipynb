```python
from __future__ import division, print_function, unicode_literals
import numpy as np
import matplotlib.pyplot as plt
```

```python
lambda_param = 1.0  # rate parameter for the exponential distribution
random_array = np.random.exponential(scale=1/lambda_param, size=(4, 4))
print("4x4 Random Array from Exponential Distribution:\n", random_array)
```

```python
large_random_array = np.random.exponential(scale=1/lambda_param, size=(100000, 1))
```

```python
uniform_samples = np.random.uniform(low=0, high=10, size=100000)
normal_samples = np.random.normal(loc=5, scale=2, size=100000)
```

```python
plt.figure(figsize=(15, 5))
```

```python
# Plot the histograms
plt.figure(figsize=(15, 5))



plt.hist(large_random_array, bins=100, density=True, alpha=0.6, color='blue')




plt.hist(uniform_samples, bins=100, density=True, alpha=0.6, color='green')


plt.hist(normal_samples, bins=100, density=True, alpha=0.6, color='red')
plt.title('Random Distribution')
plt.xlabel('Value')
plt.ylabel('Density')


plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Generate x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create meshgrid
X, Y = np.meshgrid(x, y)

# Compute Z values
Z = X**2 + Y**2

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Add labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set title
ax.set_title('3D plot of Z = X^2 + Y^2')

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Download the dataset
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv

# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin1')

# Extract relevant features
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
data = df[features]

# Calculate Pearson correlation
pearson_corr = data.corr(method='pearson')

# Calculate Spearman rank correlation
spearman_corr = data.corr(method='spearman')

# Plot heatmaps
fig, axes = plt.subplots(1, 2, figsize=(20, 8))

# Pearson correlation heatmap
sns.heatmap(pearson_corr, annot=True, fmt='.2f', cmap='coolwarm', ax=axes[0])
axes[0].set_title('Pearson Correlation Heatmap')

# Spearman rank correlation heatmap
sns.heatmap(spearman_corr, annot=True, fmt='.2f', cmap='coolwarm', ax=axes[1])
axes[1].set_title('Spearman Rank Correlation Heatmap')

plt.show()
```
