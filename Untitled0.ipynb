```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


```

```python
from google.colab import files


file_path = '/content/drive/MyDrive/Au_nanoparticle_dataset.csv'
df = pd.read_csv(file_path)
```

```python
df
```

```python

filtered_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]


print(filtered_df.head(20))


statistics = filtered_df.describe().loc[['mean', 'std', '25%', '50%', '75%']]
print(statistics)


fig, axes = plt.subplots(1, 4, figsize=(20, 5))
filtered_df.hist(['N_total', 'N_bulk', 'N_surface', 'R_avg'], bins=30, ax=axes)
plt.tight_layout()
plt.show()


sns.pairplot(filtered_df)
plt.show()


g = sns.pairplot(filtered_df, diag_kind="kde")
g.map_lower(sns.kdeplot, cmap="Blues_d")
plt.show()
```
